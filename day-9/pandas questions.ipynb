```python
# Your task is to find any employees who have overlapping project dates.
import pandas as pd

data = {
    'Emp_ID': [1, 1, 2, 2, 3, 3],
    'Project': ['Proj_A', 'Proj_B', 'Proj_C', 'Proj_D', 'Proj_E', 'Proj_F'],
    'Start_Date': ['2026-01-01', '2026-01-15', '2026-02-01', '2026-03-01', '2026-04-01', '2026-05-01'],
    'End_Date':   ['2026-01-20', '2026-02-15', '2026-02-20', '2026-04-01', '2026-04-15', '2026-06-01']
}

df_projects = pd.DataFrame(data)
# Converting to datetime objects to help you out
df_projects['Start_Date'] = pd.to_datetime(df_projects['Start_Date'])
# Note: It's 2026 currently, so the dates match the current year context.
df_projects['End_Date'] = pd.to_datetime(df_projects['End_Date'])
```

```python
df_projects
```

```python
# 1. Sort by Employee and Start Date
df_sorted = df_projects.sort_values(by=['Emp_ID', 'Start_Date'])

# 2. Get the previous project's End Date for each employee
df_sorted['Prev_End_Date'] = df_sorted.groupby('Emp_ID')['End_Date'].shift(1)

# 3. Check for overlaps (Current Start <= Previous End)
overlap_condition = df_sorted['Start_Date'] <= df_sorted['Prev_End_Date']

# 4. Filter the dataframe and get unique Employee IDs
overlapping_employees = df_sorted[overlap_condition]

print("Employees with schedule conflicts:", overlapping_employees)
```

```python
overlapping_employees
```

```python
import pandas as pd

data = {
    'Week': [1, 2, 3, 4, 5],
    'Revenue': [10000, 12000, 11000, 15000, 14500]
}

df_revenue = pd.DataFrame(data)
```

```python
# You need to calculate the absolute difference in revenue compared to the previous week, as well as the percentage growth week-over-week.
df_revenue['Previous_Revenue'] = df_revenue['Revenue'].shift(1)
df_revenue['Difference'] = df_revenue['Revenue'] - df_revenue['Previous_Revenue']
df_revenue['Percentage_Growth'] = (df_revenue['Revenue'] - df_revenue['Previous_Revenue']) / df_revenue['Previous_Revenue'] * 100
```
